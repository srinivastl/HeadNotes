## Importing libraries

```python
import pandas as pd
import fitz # PyMuPDF
import re
```

## Extracting HeadNotes from the pdfs

```python
df=pd.DataFrame()
def extract_text_between_headings(pdf_path, start_heading, end_heading):
    try:
        doc = fitz.open(pdf_path)
    except FileNotFoundError:
        print(f"The file {pdf_path} was not found.")
        return ""
    except Exception as e:
        print(f"An error occurred while opening the file: {e}")
        return ""

    text = ""
    extracting = False
    for page_num in range(len(doc)):
        page = doc[page_num]
        page_text = page.get_text("text")
        lines = page_text.split('\n')
        for line in lines:
            if end_heading in line and extracting:
                extracting = False
                break  # Stop extracting after the end heading
            if (not re.match(r'^\d|http', line)) and line!="SUPREME COURT OF INDIA" and extracting and (not re.match(r'Page', line)) and len(line)!=0 :
               text += line + '\n'
            if start_heading in line:
                extracting = True
    return text.strip()
HeadNote=[]
nl=[]
for j in range(1,12855):
  pdf_path = str(j)+'.pdf'
  start_heading = 'HEADNOTE'
  end_heading = 'JUDGMENT'
  extracted_text = extract_text_between_headings(pdf_path, start_heading, end_heading)
  if(len(extracted_text)!=0):
    nl.append(j)
    HeadNote.append(extracted_text)

df["HeadNote"]=HeadNote
```

## Extracting Judgments from the pdf

```python

def extract_text_between_headings(pdf_path, start_heading):
    try:
        # Open the PDF file
        doc = fitz.open(pdf_path)
    except FileNotFoundError:
        print(f"The file {pdf_path} was not found.")
        return ""
    except Exception as e:
        print(f"An error occurred while opening the file: {e}")
        return ""
    text = ""
    extracting = False

    # Iterate through the pages
    for page_num in range(len(doc)):
        page = doc[page_num]
        page_text = page.get_text("text")

        # Split the page text into lines
        lines = page_text.split('\n')

        for line in lines:
            if (not re.match(r'^\d|http', line)) and line!="SUPREME COURT OF INDIA" and extracting and (not re.match(r'Page', line)) and len(line)!=0 :
               text += line + '\n'
            if start_heading == line:
                extracting = True
    return text.strip()


judgment=[]
for i in range(1,12855):
  pdf_path = str(i)+'.pdf'
  start_heading = 'JUDGMENT:'
  extracted_text = extract_text_between_headings(pdf_path, start_heading)
  if(len(extracted_text)!=0) and i  in nl :
    judgment.append(extracted_text)

df["Judgment"]=judgment
```

## counting no of words in the headnotes and judgments using split

```python
def count_words_in_dataset(df):
    h_w=[]
    J_w=[]
    for index, row in df.iterrows():
        # Count words in headnotes and judgments
        headnote_words = len(str(row['HeadNote']).split())
        judgment_words = len(str(row['Judgment']).split())
        h_w.append(headnote_words)
        J_w.append(judgment_words)

    return h_w,J_w


h_w,J_w = count_words_in_dataset()

df["h_words"]=h_w
df['j_words']=J_w
```

### converting into dataset(CSV)

```python
df.to_csv("dataset.csv",index=False)
```
